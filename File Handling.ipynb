```python
import csv
```

```python
countries = [
    ['Name', 'Destination', 'Price'],
    ['USA', 'New York', '$10,000'],
    ['UK', 'London', '$12,000'],
    ['Italy', 'Rome', '$9,000'],
    ['Nepal', 'Pokhara', '$3,000']
]
```

```python
# CSV -> open(file_name, operation, newline='')
with open('countries.csv', 'w', newline='') as file: # create file configuration
    writer = csv.writer(file) # create actual csv file
    writer.writerows(countries) # insert data into csv file
```

```python
with open('countries.csv', 'r') as file: # File configuration read
    reader = csv.reader(file) # Get Data from file and Store it
    for row in reader: # Prints data
        print(row)
```

```python
new_countries = [
    ['Canada', 'Ottawa', '$13,000'],
    ['Austrailia', 'Sydney', '$10,000']
]

with open('countries.csv', 'a', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(new_countries)
```

```python
personal_details = [
    {
        'Name': 'Bob',
        'Age': 19,
        'Qualification': 'High School',
        'Address': 'California'
    },
    {
        'Name': 'Charles',
        'Age': 25,
        'Qualification': 'Bachelors',
        'Address': 'New York'
    },
    {
        'Name': 'James',
        'Age': 30,
        'Qualification': 'Masters',
        'Address': 'New Jersy'
    },
    {
        'Name': 'Bill',
        'Age': 50,
        'Qualification': 'Bachelors',
        'Address': 'Queens'
    },
    {
        'Name': 'Sadie',
        'Age': 35,
        'Qualification': 'PhD Doctrate',
        'Address': 'Brooklyn'
    },
]
```

```python
with open('personal_details.csv', 'w', newline='') as file: # File configuration
    columns = ['Name', 'Age', 'Qualification', 'Address'] # Specified Column Names
    writer = csv.DictWriter(file, fieldnames=columns) # Create file and initialize header names
    writer.writeheader() # Create header
    writer.writerows(personal_details) # insert data
```

```python
with open('personal_details.csv', 'r') as file:
    reader = csv.DictReader(file)
    for idx, row in enumerate(reader):
        print(f'{idx+1}.')
        print('-'*20)
        for key, val in row.items():
            print(f'{key}: {val}')
        print('-'*20)
```

## JSON (Javascript Object Notation)

```python
import json
```

```python
# dump -> Value write, Load -> Value read
personal_details_json = [
    {
        'Name': 'Bob',
        'Age': 19,
        'Qualification': 'High School',
        'Address': 'California'
    },
    {
        'Name': 'Charles',
        'Age': 25,
        'Qualification': 'Bachelors',
        'Address': 'New York'
    },
    {
        'Name': 'James',
        'Age': 30,
        'Qualification': 'Masters',
        'Address': 'New Jersy'
    },
    {
        'Name': 'Bill',
        'Age': 50,
        'Qualification': 'Bachelors',
        'Address': 'Queens'
    },
    {
        'Name': 'Sadie',
        'Age': 35,
        'Qualification': 'PhD Doctrate',
        'Address': 'Brooklyn'
    },
]
```

```python
with open('pd_info.json', 'w') as file:
    json.dump(personal_details_json, file, indent=4)
```

```python
with open('pd_info.json', 'r') as file:
    data = json.load(file)
    for row in data:
        print('-'*15)
        for idx, val in row.items():
            print(f'{idx}: {val}')
        print('-'*15)
```
