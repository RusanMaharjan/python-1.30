```python
def check_grade(sub1, sub2, sub3, sub4, sub5):
    total_marks = sub1 + sub2 + sub3 + sub4 + sub5
    print(f'\nTotal marks of 5 subjects: {total_marks}')
    
    marks = total_marks/5
    print(f'\nAverage marks of student: {marks}')
    
    if marks >= 80 and marks <= 100: # False  True -> False
        print('Grade A')
    elif marks >= 60 and marks < 80: # 60 - 79: False   True  -> False
        print('Grade B')
    elif marks >= 50 and marks < 60: # 50 - 59: True   True -> True
        print('Grade C')
    elif marks >= 40 and marks < 50:
        print('Grade D')
    else:
        print('Failed')
```

```python
check_grade(10, 50, 70, 40, 20)
```

### Default Argument and Keyword Argument

```python
def employee_details(fullName='Unknown Name', address='Unknown Location', 
                     jobPosition='Unknown Position', jobType='Undefined Type'):
    print(f'Employee Full Name: {fullName}')
    print(f'Employee Address: {address}')
    print(f'Employee Job Position: {jobPosition}')
    print(f'Employee Job Type: {jobType}')
```

```python
employee_details('UK')
```

```python
employee_details(address='Uk', jobType='Part-Time')
```

## All Arguments and All Keyword Arguments

```python
'''
All Arguments (*args)
All Keyword Arguments (**kwargs)
'''
```

```python
def product_details(*args):
    print(args)
    for data in args:
        print(f'Details of Mobile: {data}')
```

```python
product_details('Samsung', 'Rs.250000', 'Ultra Series')
```

```python
def product_details_kwargs(**kwargs):
    print(kwargs)
    for key, val in kwargs.items():
        print(f'{key} of Mobile: {val}')
```

```python
product_details_kwargs(Brand = 'Samsung', Price = 'Rs.200000', Model = 'Ultra Series')
```

## Nested Functions

```python
def add(x, y):
    print(f'Addition of {x} and {y} is: {x + y}')

def sub(x, y):
    print(f'Subtraction of {x} and {y} is: {x - y}')

def mul(x, y):
    print(f'Multiplication of {x} and {y} is: {x * y}')

def div(x, y):
    print(f'Division of {x} and {y} is: {x / y}')
```

```python
def calculate():
    x = int(input('Enter first number: '))
    y = int(input('Enter second number: '))
    add(x, y)
    sub(x, y)
    mul(x, y)
    div(x, y)
```

```python
calculate()
```

```python
def add(x, y):
    return f'Addition of {x} and {y} is: {x + y}'

def sub(x, y):
    return f'Subtraction of {x} and {y} is: {x - y}'

def mul(x, y):
    return f'Multiplication of {x} and {y} is: {x * y}'

def div(x, y):
    return f'Division of {x} and {y} is: {x / y}'
```

```python
def calculate(x, y):
    add(x, y)
    sub(x, y)
    mul(x, y)
    div(x, y)
    return add(x, y), sub(x, y), mul(x, y), div(x, y)
```

```python
calculate(6, 7)
```

## Lambda Functions

```python
def add_data(a, b):
    add = a + b
    print(add)
```

```python
add_data(1, 2)
```

```python
# function_name = lambda arg1, arg2 : condition / print
```

```python
add_data = lambda a, b : a + b
add_data(5, 6)
```

```python
odd_even = lambda num : "Even Number" if num % 2 == 0 else "Odd Number"
odd_even(7)
```
