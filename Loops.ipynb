```python
# Remaining Attempts
balance = 60000
pin = 1234

for i in range(3):
    user_pin = int(input('Enter your pin number: '))
    
    if user_pin == pin:
        user_balance = int(input('Enter your withdraw balance amount: '))
        if user_balance < balance:
            if user_balance % 1000 == 0:
                print(f'Rs.{user_balance} has been withdrawn from A/C 1234567890.')
                remaining = balance - user_balance
                print(f'Remaining Balance: Rs.{remaining}')
                break
            else:
                print('Amount must be multiple of 1000.')
        else:
            print('Withdraw Balance amount must be less than actual balance.')
    else:
        print('Incorrect Pin Number.')
else:
    print('Card Blocked. Kindly contact nearest branch.')
```

### While Loop

```python
# Guess the Number game 
# Write a program using while loop where: 
# Any number is defined at first: eg., secret_value = 7 
# The user keeps guessing until the correct number is guessed. 
```

```python
secret_number = 7
number = 0

while number != secret_number: #   != 7 -> False
    number = int(input('Enter a number: '))

    if number == secret_number:
        print('You choosed correct value.')
    else:
        print('Wrong value. Try again!')
```

```python
# Create a food order program: 
# Where user can order any items, they want to order 
# Loop should run until user choose number 5 to exit. 
# 1. Momo 
# 2. Chowmin 
# 3. Pizza 
# 4. Drinks 
# 5. Exit
```

```python
order_num = 0

while order_num != 5:
    order_num = int(input('Enter a number: '))
    
    match order_num:
        case 1:
            print('You have ordered Momo.')
        case 2:
            print('You have ordered Chowmin.')
        case 3:
            print('You have ordered Pizza.')
        case 4:
            print('You have ordered Drinks')
        case 5:
            print('Thank You for choosing us.')
        case _:
            print('Invalid Number.')
```

```python
# Remaining Attemp: 2
# Invalid Username or password.

# Remaining attempt: 1
# Invalid username or Password.

# Remaining attempt: 0
# Invalid username or password.

# Account Blocked.
```

```python
uname = 'abc@gmail.com'
pwd = '12345678'

username = None
password = None

attempts = 3

while username != uname and password != pwd:
    attempts -= 1 # attempts = attempts - 1
    print(f'Remaining Attempts: {attempts}')
    username = input('Enter username: ')
    password = input('Enter password: ')

    if username == uname and password == pwd:
        print('Login Successfull!!\n')
    else:
        print('Login Failed.\n')

    if attempts == 0:
        print('Account Blocked!!')
        break
```
