# Error Handling

```python
'''
try, except, else, finally, Custom Exception
'''
```

```python
try:
    num1 = int(input('Enter first number: '))
    num2 = int(input('Enter second number: '))
    
    num3 = num1 / num2
    print(num3)
except ZeroDivisionError:
    print('Number cannot be divisible by 0.')
except ValueError:
    print('Only number integer values are accepted in input field.')
```

```python
try:
    with open('text_fifdfe.txt', 'r') as file:
        content = file.read()
        print(content)
except FileNotFoundError as e:
    print(e)
```

```python
try:
    file_name = input('Enter text file name to read: ')
    with open(f'{file_name}.txt', 'r') as file:
        content = file.read()
        print(content)
except FileNotFoundError as e:
    print(e)
```

```python
try:
    file_name = input('Enter text file name to read: ')
    with open(f'{file_name}.txt', 'r') as file:
        content = file.read()
except FileNotFoundError as e:
    print(e)
else:
    print(content)
finally:
    print('Program ended successfully.')
```

```python
## Age Exception
```

```python
class AgeExceptionError(Exception):
    pass
```

```python
def check_age(age):
    if age >= 18:
        print('You are eligible to vote!!')
    else:
        raise AgeExceptionError(f'Invalid age {age}. Age must be more than or equals to 18.')
```

```python
try:
    age = int(input('Enter your age: '))
    check_age(age)
except AgeExceptionError as e:
    print(e)
except ValueError:
    print(f'Age value must be integer.')
```
