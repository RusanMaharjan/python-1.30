```python
first_name = 'Ram'
last_name = 'Bahadur'
age = '30'
qualification = 'Bachelors'
```

```python
# Employee Name: Ram Bahadur
# Employee Age: 30
# Employee Qualification: Bachelors
```

```python
print(f'Employee Name: {first_name} {last_name}')
# \n -> New line
# \t -> Tab
print(f'Employee Age: {age}\nEmployee Qualification: {qualification}')
```

```python
'''
Operators
--------------
1. Arithmetic Operator
    -> +, -, *, /, % (reminder / modulus), //(float division)
2. Logical Operator
    -> AND, OR, NOT
3. Comparison Operator
    -> =, ==, !=, <, >, <=, >=
'''
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

add = num1 + num2
sub = num1 - num2
mul = num1 * num2
div = num1 / num2
rem = num1 % num2
fd = num1 // num2
```

```python
print(add)
print(sub)
print(mul)
print(div)
print(rem)
print(fd)
```

```python
num3 = 10
```
