### Student Grade Calculator

```python
marks = float(input('Enter your marks: '))
# A, B, C, D, Fail
if marks >= 80 and marks <= 100: # False  True -> False
    print('Grade A')
elif marks >= 60 and marks < 80: # 60 - 79: False   True  -> False
    print('Grade B')
elif marks >= 50 and marks < 60: # 50 - 59: True   True -> True
    print('Grade C')
elif marks >= 40 and marks < 50:
    print('Grade D')
else:
    print('Failed')
```

```python
sub1 = float(input('Enter your marks for subject 1: '))
sub2 = float(input('Enter your marks for subject 2: '))
sub3 = float(input('Enter your marks for subject 3: '))
sub4 = float(input('Enter your marks for subject 4: '))
sub5 = float(input('Enter your marks for subject 5: '))

# \n -> new line
# \t -> tab

total_marks = sub1 + sub2 + sub3 + sub4 + sub5
print(f'\nTotal marks of 5 subjects: {total_marks}')

marks = total_marks/5
print(f'\nAverage marks of student: {marks}')

if marks >= 80 and marks <= 100: # False  True -> False
    print('Grade A')
elif marks >= 60 and marks < 80: # 60 - 79: False   True  -> False
    print('Grade B')
elif marks >= 50 and marks < 60: # 50 - 59: True   True -> True
    print('Grade C')
elif marks >= 40 and marks < 50:
    print('Grade D')
else:
    print('Failed')
```

### ATM

```python
# 1. pin correct
# 2. withdraw amount less than balance amount
current_balance = 50000
user_pin = 1234

pin = int(input('Enter your pin number: '))
balance = int(input('Enter balance to withdraw: '))

if pin == user_pin and balance < current_balance:
    # Calculation of remaining balance
    remaining_balance = current_balance - balance
    # print balance
    print(f'Rs.{balance} is withdrawn from A/C 12345678.')
    print(f'Remaining Balance: Rs.{remaining_balance}')
    
elif pin != user_pin and balance < current_balance:
    print('Incorrect Pin!!')
    
elif pin == user_pin and balance > current_balance:
    print('Insufficient Balance!!')
else:
    print('Incorrect Pin Number or Insufficient Balance.')
```

```python
# pin correct and balance insufficient
# -> Insufficient Balance

# pin incorrect and balance sufficient
# -> incorrect pin number
```

```python
# Login authentication system
# User -> Username and password
# Login
# User not found / Incorrect Password

uname = 'abc@gmail.com'
pwd = '1234hh'

username = input('Enter your username: ')
password = input('Enter your password: ')

if username == uname and password == pwd:
    print('Login Successfull!!')
elif username == uname and password != pwd:
    print('Login Failed. Incorrect Password!!')
elif username != uname and password == pwd:
    print('Login Failed. Username not found.')
else:
    print('Login Failed. Username or Password Incorrect!!')
```

```python
'''
calculate tax based on annual income 
<=250000 then no tax 
250001 - 500000 then 5% tax 
500001 - 1000000 then 20% 
more than 1000000 then 30% 

output
------------
Annual Income: 900000
Taxable Amount (20%): tax amount
Remaining Amount: 
'''
```

```python
annual_income = int(input('Enter annual income: '))

if annual_income > 0 and annual_income <= 250000:
    print(f'Annual Income: Rs.{annual_income}')
    print(f'Taxable Amount: Rs.0')
    print(f'Remaining Amount: Rs.{annual_income}')
    
elif annual_income > 250000 and annual_income <= 500000:
    # print('Annual Income: Rs.',annual_income)
    print(f'Annual Income: Rs.{annual_income}')
    # tax = (annual_income * 5)/100
    tax = annual_income * 0.05
    print(f'Taxable Amount: Rs.{tax}')
    remaining_amount = annual_income - tax
    print(f'Remaining Amount(5%): Rs.{remaining_amount}')
    
elif annual_income > 500000 and annual_income <= 1000000:
    print(f'Annual Income: Rs.{annual_income}')
    # tax = (annual_income * 5)/100
    tax = annual_income * 0.20
    print(f'Taxable Amount(20%): Rs.{tax}')
    remaining_amount = annual_income - tax
    print(f'Remaining Amount: Rs.{remaining_amount}')
    
elif annual_income > 1000000:
    print(f'Annual Income: Rs.{annual_income}')
    # tax = (annual_income * 5)/100
    tax = annual_income * 0.30
    print(f'Taxable Amount(30%): Rs.{tax}')
    remaining_amount = annual_income - tax
    print(f'Remaining Amount: Rs.{remaining_amount}')
    
else:
    print('Amount must be more thn 0 or Positive integer.')
```

## Nested If

```python
# A student is eligible for an exam if: 
# Attendance >= 75 
# internal marks >= 40 
# print exact reason if not eligible 

attendance = int(input('Enter attendnace: '))
internal_marks = float(input('Enter marks: '))

if attendance >= 75 and internal_marks >= 40:
    print('Eligible for Exam.')
elif attendance < 75 and internal_marks >= 40:
    print('Not eligible for Exam. Attendance Requirement does not meet.')
elif attendance >= 75 and internal_marks < 40:
    print('Not eligible for Exam. Internal Marks Requirement does not meet.')
else:
    print('Not eligiblie for Exam.')
```

```python
attendance = int(input('Enter attendnace: '))

if attendance >= 75: # False
    internal_marks = float(input('Enter marks: '))
    if internal_marks >= 40: # True
        print('Eligible for Exam.')
    else:
        print('Not eligible for Exam. Internal Marks Requirement does not meet.')
else:
    print('Not eligible for Exam. Attendance Requirement does not meet.')
```

```python
# balance = 60000

# pin correct

# withdraw amount -> 10000

# withdraw < balance

# 1000s -> 1000, 2000, 3000

# withdraw amount->
# remain balance ->
```

```python
balance = 60000
pin = 1234

user_pin = int(input('Enter your pin number: '))

if user_pin == pin:
    user_balance = int(input('Enter your withdraw balance amount: '))
    if user_balance < balance:
        if user_balance % 1000 == 0:
            print(f'Rs.{user_balance} has been withdrawn from A/C 1234567890.')
            remaining = balance - user_balance
            print(f'Remaining Balance: Rs.{remaining}')
        else:
            print('Amount must be multiple of 1000.')
    else:
        print('Withdraw Balance amount must be less than actual balance.')
else:
    print('Incorrect Pin Number.')
```

```python
# match case

print('Restaurant Menu')
print('1. Breakfast')
print('2. Lunch')
print('3. Snaks')
print('4. Dinner')
print('5. Drinks')
print('6. Exit')

order_num = int(input('Enter order number (1 - 6): '))

match order_num:
    case 1:
        print('You have ordered breakfast.')
    case 2:
        print('You have ordered Lunch.')
    case 3:
        print('You have ordered Snaks.')
    case 4:
        print('You have ordered Dinner.')
    case 5:
        print('You have ordered Drinks.')
    case 6:
        print('Thank You for choosing us.')
    case _: # default case
        print('Invalid Number. Choose number from 1 - 6.')
```
