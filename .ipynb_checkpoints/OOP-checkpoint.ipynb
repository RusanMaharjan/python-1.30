```python
# __init__() -> Constructor
c_name = 'ABC Company'
class Employees:
    def __init__(self, name, empType, empPosition, empSalary):
        self.company_name = c_name
        self.company_address = 'Kathmandu'
        self.name = name
        self.empType = empType
        self.empPosition = empPosition
        self.empSalary = empSalary

    def print_details(self):
        print(self.company_name)
        print(self.company_address)
        print(self.name)
        print(self.empType)
        print(self.empPosition)
        print(self.empSalary)
```

```python
emp_name = input('Enter your name: ')
e = Employees(emp_name, 'Part-Time', 'Junior Python Developer', '$20000')
e.print_details()
```

```python
# Lab Question 3: Professional Profile (Mixed Attributes) Goal: Understand the difference 
# between shared class data and unique object data. 
# 1. Define a class Employee. 
# 2. Set two class attributes: company_name = "TechCorp" and department = 
# "Development". 
# 3. Use __init__ to accept instance-specific data: name, designation, and salary. 
# 4. Create a method display_info() that returns a summary string containing both the 
# shared company name and the specific employee's name. 
# 5. Create an instance and print the result of the display_info() method. 
```

```python
class Employee:
    def __init__(self, name, designation, salary):
        self.name = name
        self.designation = designation
        self.salary = salary
        self.company_name = 'TechCorp'
        self.department = 'Development'

    def display_info(self):
        print(self.company_name)
        print(self.department)
        print(self.name)
        print(self.designation)
        print(self.salary)
```

```python
e = Employee('Ram', 'Software Engineer', '25000')
e.display_info()
```

## Inheritance

```python
'''
Types of Inheritance
-----------------------
1. Simple Inheritance
2. Multiple Inheritance
3. Multi-Level Inheritance
'''

'''
1. Child class inherit/extends parent class
abstract class -> (School) -> implements
normal class -> (School) -> extends / inherit
'''
```

```python
# School (one)   (multiple) Teacher
# ENUM/.env File
SCHOOL_NAME = 'Abc School'
```

```python
# Parent Class School
class School:
    def __init__(self, school_address, contact_number):
        self.school_name = SCHOOL_NAME
        self.school_address = school_address
        self.contact_number = contact_number

    def school_details(self):
        print(f'Name of School: {self.school_name}')
        print(f'Address of School: {self.school_address}')
        print(f'School Contact: {self.contact_number}')
```

```python
# child class
class Teacher(School):
    def __init__(self, school_address, contact_number, teacher_id, teacher_name, teacher_contact, 
                 number_of_subject, teacher_qualification):
        # to inherit constructor of parent class -> super() keyword is used.
        super().__init__(school_address, contact_number)
        self.teacher_id = teacher_id
        self.teacher_name = teacher_name
        self.teacher_contact = teacher_contact
        self.number_of_subject = number_of_subject
        self.teacher_qualification = teacher_qualification

    def teacher_display_info(self):
        print(f'Teacher ID: {self.teacher_id}')
        print(f'Name of Teacher: {self.teacher_name}')
        print(f'Teacher Contact: {self.teacher_contact}')
        print(f'Number of Subjects: {self.number_of_subject}')
        print(f'Teacher Qualification: {self.teacher_qualification}')
```

```python
t = Teacher('Kathamndu', '01-4235635', 1, 'Ram Bahadur', 9876543210, 5, 'Masters')
```

```python
t.school_details()
t.teacher_display_info()
```

```python
class Student(School):
    def __init__(self, school_address, contact_number, student_name, student_address, student_contact, student_grade, student_section):
        super().__init__(school_address, contact_number)
        self.student_name = student_name
        self.student_address = student_address
        self.student_contact = student_contact
        self.student_grade = student_grade
        self.student_section = student_section

    def student_display_info(self):
        print(f'Name of Student: {self.student_name}')
        print(f'Student Address: {self.student_address}')
        print(f'Student Contact: {self.student_contact}')
        print(f'Student Grade: {self.student_grade}')
        print(f'Section: {self.student_section}')
```

```python
s = Student('Kathmandu', '1234567890', 'Bill Watson', 'Manhatten', '9876543210', 'B+', 'C7')
```

```python
s.school_details()
s.student_display_info()
```

### Multi-Level Inheritance

```python
# parent class
class Product:
    def __init__(self, product_name, product_desc, product_price, product_category, product_quantity):
        self.product_name = product_name
        self.product_desc = product_desc
        self.product_price = product_price
        self.product_category = product_category
        self.product_quantity = product_quantity

    def product_details(self):
        print(f'Name of Product: {self.product_name}')
        print(f'Product Description: {self.product_desc}')
        print(f'Product Price: {self.product_price}')
        print(f'Product Category: {self.product_category}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Total Price: Rs.{self.product_price * self.product_quantity}')
```

```python
class Order(Product):
    def __init__(self, order_id, order_date, order_time, order_destination, delivery_charge,
                product_name, product_desc, product_price, product_category, product_quantity):
        super().__init__(product_name, product_desc, product_price, product_category, product_quantity)
        self.order_id = order_id
        self.order_date = order_date
        self.order_time = order_time
        self.order_destination = order_destination
        self.delivery_charge = delivery_charge

    def order_details(self):
        print(f'Order Id: {self.order_id}')
        print(f'Order Date: {self.order_date}')
        print(f'Order Time: {self.order_time}')
        print(f'Order Destination: {self.order_destination}')
        print(f'Delivery Charge: {self.delivery_charge}')
```

```python
class Customer(Order):
    def __init__(self, customer_name, customer_address, customer_contact,
                order_id, order_date, order_time, order_destination, delivery_charge,
                product_name, product_desc, product_price, product_category, product_quantity):
        super().__init__(order_id, order_date, order_time, order_destination, delivery_charge,
                product_name, product_desc, product_price, product_category, product_quantity)
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_contact = customer_contact

    def customer_details(self):
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Contact: {self.customer_contact}')
```

```python
c = Customer('Bob', 'Jhapa', '9876543210', 10, '20th May, 2026', '2:00 PM', 'Biratnagar', 400,
            'Bottle', 'Black stainless bottle. 24 Hrs Heat', 500, 'Utensils', 4)
```

```python
c.customer_name
```

```python
c.customer_details()
```

```python
c.order_details()
```

```python
c.product_details()
```

## Encapsulation

```python
'''
Data Hiding
--------------
Access Modifiers
1. Private -> self.__name
2. Public -> self.name
3. Protected -> self._name
'''
```

```python
class Bank:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print(f'Deposit Amount: Rs.{amount}')
        print(f'Total Balance: {self.__balance}')

    def withdraw(self, amount):
        self.__balance -= amount
        print(f'Withdraw Amount: Rs.{amount}')
        print(f'Total Balance: {self.__balance}')
```

```python
b = Bank(50000)
```

```python
b.deposit(3000)
```

```python
b.withdraw(2000)
```

```python
class Product:
    def __init__(self, p_name, p_price):
        self.__p_name = p_name
        self.__p_price = p_price
        self.__cost_price = 299

    def details(self):
        print(f'Product Name: {self.__p_name}')
        print(f'Productt Price: {self.__p_price}')

    def __profit(self):
        print(f'Product Profit: Rs.{self.__p_price - self.__cost_price}')

    def show_profit(self):
        self.__profit()
```

```python
p = Product('ABC', 567)
p.show_profit()
```

```python
class Customer(Product):
    def __init__(self, c_name, c_phone, p_name, p_price):
        super().__init__(p_name, p_price)
        self.__c_name = c_name
        self.__c_phone = c_phone

    def customer_details(self):
        print(f'Customer Name: {self.__c_name}')
        print(f'Customer Phone: {self.__c_phone}')
```

```python
c = Customer('Bob', '1234567890', 'Mobile', 34569)
```

```python
c.details()
c.customer_details()
c.show_profit()
```

## Polymorphism

#### Message Gateway

```python
from datetime import datetime
class Messenger:
    def message(self, send_by, send_to, message):
        print(f'{send_to} - Facebook(1)')
        print(f'{send_by} send you a message.')
        print(f'{message}')
        print(f'Time: {datetime.now()}')

class Viber:
    def message(self, send_by, send_to, message):
        print(f'{send_to} - Viber(1)')
        print(f'{send_by} send you a message.')
        print(f'{message}')
        print(f'Time: {datetime.now()}')

class WhatsApp:
    def message(self, send_by, send_to, message):
        print(f'{send_to} - WhatsApp(1)')
        print(f'{send_by} send you a message.')
        print(f'{message}')
        print(f'Time: {datetime.now()}')

class Slack:
    def message(self, send_by, send_to, message):
        print(f'{send_to} - Slack(1)')
        print(f'{send_by} send you a message.')
        print(f'{message}')
        print(f'Time: {datetime.now()}')        
```

```python
def send_message(message_obj, by, to, msg):
    message_obj.message(by, to, msg)
```

```python
send_message(Messenger(), 'Ram', 'Shyam', 'Hello!!')
```
