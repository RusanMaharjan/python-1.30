```python
'''
Python Data Structure
----------------------
1. List -> []
2. Dictionary -> {}
3. Sets -> {}
4. Tuple -> ()

a. Mutable Data Structure
    -> Data Structure in which we can perform value add/update/delete/read operation. 
    -> List, Dictionary, Sets

b. Immutable Data Structure
    -> Data Structure in which we only can perform read operation.
    -> Tuple
'''
```

## Dictionary

```python
# JSON -> Javascript Object Notation
```

```python
# {}
'''
- Dictionary contains key value pair.
- Dictionary can contain list.
'''
```

```python
person = {
    'Name': 'Ram',
    'Age': 30,
    'Address': 'Kathmandu',
    'Qualification': 'Masters'
}
person
```

```python
personal_details = {
    'Name': ['Bob', 'Stuart', 'John'],
    'Age': [30, 20, 50],
    'Address': ['Italy', 'USA', 'UK'],
    'Qualification': ['Bachelors', 'Masters', 'Bachelors']
}
personal_details
```

```python
personal_details['Name']
```

```python
personal_details['Age']
```

```python
# .keys(), .values(), .items()
for key in personal_details.keys():
    print(key)
```

```python
for val in personal_details.values():
    print(val)
```

```python
for key, val in personal_details.items():
    print(f'{key}: {val}')
```

```python
for key, val in personal_details.items():
    print(f'\n{key}')
    print('-'*10)
    for data in val:
        print(data)
```

### Add value in dictionary

```python
personal_details['Phone'] = [1234567890, 987654321, 567890432]
personal_details
```

```python
personal_details['Name'].append('Aurthr')
```

```python
personal_details
```

### Update Data

```python
# personal_details.update({'Qualification': 'Bachelors'})
# personal_details
```

```python
personal_details['Qualification'][1] = 'Bachelors'
```

```python
personal_details
```

### Delete data

```python
'''
1. Delete -> del -> key
2. pop(key)
3. clear()
'''
```

```python
# personal_details.remove('Age')
del personal_details['Age']
```

```python
personal_details.pop('Name')
```

```python
personal_details
```

```python
personal_details.clear()
```

```python
personal_details
```

## Sets

```python
# List, Dictionary, Tuple -> Index or Keys
# {}, set()
```

```python
set_a = {1, 2, 3, 4, 5}
print(set_a)
```

```python
type(set_a)
```

```python
numbers = [1, 2, 3, 4, 5]
set_b = set(numbers)
set_b
```

```python
num = [1, 0, 4, 3, 2, 4, 5, 5, 5, 5, 5, 6, 8, 6, 7, 2, 1, 4, 3, 9, 0, 0, 0, 0]
num_set = set(num)
num_set
```

```python
num_set.add(0.5)
num_set
```

```python
num_set.update([1000, 100, 99, 105, 104, 104, 104, 105])
num_set
```

```python
'''
1. remove(value)
2. discard(value)
3. pop()
'''
```

```python
num_set
```

```python
num_set.remove(3)
num_set
```

```python
num_set.discard(3)
num_set
```

```python
num_set.pop()
```
