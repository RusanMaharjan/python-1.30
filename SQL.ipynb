```python
import sqlite3
```

```python
conn = sqlite3.connect('sql1.db')
```

```python
cursor = conn.cursor()
```

## Create Table

```python
try:
    cursor.execute('''
        Create Table Customers (
        	Customer_Id integer primary key autoincrement,
        	first_name text not null,
        	last_name text not null,
        	email text not null,
        	phone text not null,
        	city text,
        	state text not null
        );
    ''')
    print('Table customer created.')
except:
    print('Table customer already exists.')
```

```python
try:
    cursor.execute('''
        Create Table Staff (
            Staff_id integer primary key autoincrement,
            first_name text not null,
            last_name text not null,
            email text not null,
            phone text not null,
            position text not null
        );
    ''')
    print('Staff table created successfully.')
except:
    print('Table staff already exists.')
```
